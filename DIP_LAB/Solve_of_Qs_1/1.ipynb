```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("flower.bmp", cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
```

```python
x = np.asarray(img)
print(f"Image shape: {img.shape}\n")
print(x)
```

```python
newX = x[::2,::2]
print(len(newX))
cv2.imwrite("newFlower.bmp", newX)
```

```python
newImg = cv2.imread("newFlower.bmp",cv2.IMREAD_GRAYSCALE)
print(f"New Image shape: {newImg.shape}")
print(f"Original Image shape: {img.shape}")
plt.subplot(1,2,2)
plt.imshow(newImg, cmap = "gray")
plt.title("Reduced Image: 256X256")
plt.subplot(1,2,1)
plt.imshow(img, cmap = "gray")
plt.title("Original Image: 512X512")
```

Histogram

```python
hx= np.zeros(256)
for i in range(len(x)):
    for j in range(len(x[i])):
        hx[x[i,j]] = hx[x[i,j]] + 1

hNx= np.zeros(256)
for i in range(len(newX)):
    for j in range(len(newX[i])):
        hNx[newX[i,j]] = hNx[newX[i,j]] + 1

plt.subplot(2,1,1)
plt.bar(range(256), hx, width=1, color = 'grey', alpha = 0.8)
plt.title("Histogram of Original Image")
plt.subplot(2,1,2)
plt.bar(range(256), hNx, width=1, color = 'grey', alpha = 0.8)
plt.title("Histogram of Reduced Image")
plt.tight_layout()
```

Reduce Intensity level

```python
arr2d = []
for i in range(len(x)):
    arr = []
    for j in range(len(x[i])):
        arr.append(x[i,j]/2)
    arr2d.append(arr)
nwIX = np.asarray(arr2d)
```

```python
cv2.imwrite("newFlowerAfterIntensityReduce.bmp", nwIX)
```

```python
newImg2 = cv2.imread("newFlowerAfterIntensityReduce.bmp",cv2.IMREAD_GRAYSCALE)
plt.subplot(1,2,2)
plt.imshow(newImg2, cmap = "gray")
plt.title("Image after Reduce Intensity")
plt.subplot(1,2,1)
plt.imshow(img, cmap = "gray")
plt.title("Original Image")
plt.tight_layout()
```

```python
hNIx= np.zeros(256)
for i in range(len(nwIX)):
    for j in range(len(nwIX[i])):
        hNIx[int(nwIX[i,j])] = hNIx[int(nwIX[i,j])] + 1

plt.subplot(2,1,1)
plt.bar(range(256), hx, width=1, color = 'grey', alpha = 0.8)
plt.title("Histogram of Original Image")
plt.subplot(2,1,2)
plt.bar(range(256), hNIx, width=1, color = 'grey', alpha = 0.8)
plt.title("Histogram of Reduced Image")
plt.tight_layout()
```

```python
img = cv2.imread("moon.bmp", cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')

x = np.asarray(img)
print(f"Image shape: {img.shape}\n")
print(x)

hx= np.zeros(256)
for i in range(len(x)):
    for j in range(len(x[i])):
        hx[x[i,j]] = hx[x[i,j]] + 1
```

```python
mx = 255
mn = 0
threshold = 127

arr2d = []
for i in range(len(x)):
    arr = []
    for j in range(len(x[i])):
        if x[i,j] > threshold:
            arr.append(mx)
        else:
            arr.append(mn)
    arr2d.append(arr)
nwIX = np.asarray(arr2d)
```

```python
cv2.imwrite("MoonAfterThresholding.bmp", nwIX)
```

```python
newImg2 = cv2.imread("MoonAfterThresholding.bmp",cv2.IMREAD_GRAYSCALE)
plt.subplot(1,2,2)
plt.imshow(newImg2, cmap = "gray")
plt.title("threshold segmentation")
plt.subplot(1,2,1)
plt.imshow(img, cmap = "gray")
plt.title("Original Image")
plt.tight_layout()
```
