Import modules

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lena.bmp', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='grey')
plt.show()
```

Adding salt and papper noise

adding some salt (white) noise

```python
noisy_img = img
row, col = img.shape
num_of_px = np.random.randint(500, 10000)
for i in range(num_of_px):
    x = np.random.randint(0,row-1)
    y = np.random.randint(0,col-1)
    noisy_img[x,y] = 255

num_of_px = np.random.randint(500, 10000)
for i in range(num_of_px):
    x = np.random.randint(0,row-1)
    y = np.random.randint(0,col-1)
    noisy_img[x,y] = 0
cv2.imwrite('noisy_img.bmp',noisy_img)
```

show noised image

```python
plt.imshow(noisy_img, cmap='grey')
plt.show()
```

Average Filter and Median Filter

Average filter (5X5)

```python
#filter
mask = np.ones([5, 5], dtype = int) 
mask = mask / 25
print(mask)

#perform convolution operation
row, col = noisy_img.shape
new_img = np.zeros([row, col])

for i in range(2,row-2):
    for j in range(2, col-2):
        tem = 0
        for ii in range(-2,3):
            for jj in range(-2,3):
                tem += mask[ii+2,jj+2] * noisy_img[i+ii,j+jj]
        new_img[i,j] = tem

new_img = new_img.astype(np.uint8)
#show the image
plt.imshow(new_img, cmap='grey')
plt.title('After using Average filter')
plt.show()
```

Median filter

```python
new_img2 = np.zeros([row, col])
for i in range(2, row-2): 
    for j in range(2, col-2): 
        temp = []
        for ii in range(-2,3):
            for jj in range(-2,3):
                temp.append(noisy_img[i+ii,j+jj])
        temp = sorted(temp)
        new_img2[i,j] = temp[12]

new_img2 = new_img2.astype(np.uint8)
plt.imshow(new_img2, cmap='grey')
plt.title("After using Median Filter")
plt.show()
```
