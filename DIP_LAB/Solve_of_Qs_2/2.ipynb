```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("moon.bmp", cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
```

```python
x = np.asarray(img)
print(f"Image shape: {img.shape}\n")
print(x)

hx= np.zeros(256)
for i in range(len(x)):
    for j in range(len(x[i])):
        hx[x[i,j]] = hx[x[i,j]] + 1
```

```python
plt.bar(range(256), hx, width=1, color = 'grey', alpha = 0.8)
plt.title("Histogram of Original Image")
```

```python
A = 100
B = 150
FX = 50

arr2d = []
for i in range(len(x)):
    arr = []
    for j in range(len(x[i])):
        if x[i,j] >= A and x[i,j] <= B:
            arr.append(x[i,j] + FX)
        else :
            arr.append(x[i,j])
    arr2d.append(arr)
nwIX = np.asarray(arr2d)
```

A

```python
cv2.imwrite("AfterChanedIntensity.bmp", nwIX)
```

```python
newImg2 = cv2.imread("AfterChanedIntensity.bmp",cv2.IMREAD_GRAYSCALE)
plt.subplot(1,2,2)
plt.imshow(newImg2, cmap = "gray")
plt.title("Image after Enhance Brightness (specific range)")
plt.subplot(1,2,1)
plt.imshow(img, cmap = "gray")
plt.title("Original Image")
plt.tight_layout()
```

```python
hNIx= np.zeros(256)
for i in range(len(nwIX)):
    for j in range(len(nwIX[i])):
        hNIx[int(nwIX[i,j])] = hNIx[int(nwIX[i,j])] + 1

plt.subplot(2,1,1)
plt.bar(range(256), hx, width=1, color = 'grey', alpha = 0.8)
plt.title("Histogram of Original Image")
plt.subplot(2,1,2)
plt.bar(range(256), hNIx, width=1, color = 'grey', alpha = 0.8)
plt.title("Histogram of Reduced Image")
plt.tight_layout()
```
