```python
import cv2
import matplotlib.pyplot as plt
from numpy import array
import numpy as np
```

```python
img = cv2.imread("addition_16.bmp", cv2.IMREAD_GRAYSCALE)
print(img.shape)
plt.imshow(img,cmap='gray')
x = np.asarray(img)
print(x)
```

```python
h = []
mn=0
mx=0
for arr in x:
    for val in arr:
        if val == 0:
            mn = mn+1
        elif val==255:
            mx = mx+1
        h.append(val)
plt.hist(h)
print(f"Number of 0 : {mn}")
print(f"Number of 255 : {mx}")
```

```python
img2 = cv2.imread("A.bmp",cv2.IMREAD_GRAYSCALE)
plt.subplot(1,2,1)
plt.imshow(img2,cmap='gray')
rotated_img = cv2.rotate(img2,cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.subplot(1,2,2)
plt.imshow(rotated_img,cmap='gray')
```
